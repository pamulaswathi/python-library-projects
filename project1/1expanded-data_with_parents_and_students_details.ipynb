```python
#drop unnamed coulmn

df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

```python
#change weekly study hour column if the wklystudyhour 5-oct hotho
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("05-Oct","5-10")
```

```python
df.head()
```

```python
#gender distribution
plt.figure(figsize=(4,4))#it look like chart
sns.countplot(data=df,x="Gender")
plt.show()
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.show()
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({
     "MathScore":"mean",
     "ReadingScore":"mean",
     "WritingScore":"mean"})
    
```

```python
print(gb1)
```

```python
sns.heatmap(gb1,annot=True)
plt.title("maritalstatus vs student score")
plt.show()
```

```python
#from the above chart we have concluded that marital stauts do not effect in students score
```

```python
sns.boxplot(data=df,x="MathScore")
plt.show()
```

```python
groupA=df.loc[(df["EthnicGroup"] == 'group A')].count()
groupB=df.loc[(df["EthnicGroup"] == 'group B')].count()
groupC=df.loc[(df["EthnicGroup"] == 'group C')].count()
groupD=df.loc[(df["EthnicGroup"] == 'group D')].count()

groupE=df.loc[(df["EthnicGroup"] == 'group E')].count()

l=["group A","group B","group C","Group D","group E"]
mylist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mylist,labels=l)
plt.title("Distribution of ethinci group")
plt.show()
```
